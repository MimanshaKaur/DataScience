```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

### Seaborn graphs
- countplot
- histplot
- lineplot
- scatterplot
- barplot
- displot
- displot(kind=kde)
- displot(kind=hist)
- displot(with hue)
- catplot

```python
sns.countplot(df,x='class')   #creating graph through seaborn library----countplot is similar to bar graph
plt.title('number of passengers in each class')   # title()--to add graph title
plt.xlabel('class')  #xlabel()-- to add labelling to x-axis
plt.ylabel('number of passengers')   #ylabel()-- to add labelling to y-axis
plt.show()
```

```python
sns.histplot(df,x='age',kde=True, color='red')    # kde helps in providing a wave fpr more clear view of growth in graph
plt.title('age distribution of passengers')
plt.xlabel('age')
plt.ylabel('number of passengers')
plt.show()
```

```python
sns.lineplot(df,x='class',y='fare', linestyle='dashed', marker='o' , color='purple')  # marker='o' is used to show meeting pts.---color used to change color of graph---linestyle is used to change line style
plt.title("fare of each class")
plt.xlabel('class')
plt.ylabel('fare')
plt.show()
```

```python
sns.scatterplot(df,x='age', y='fare', color='red')
plt.title('fare of each age')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
sns.displot(df, x='age', y='class', color='red')
plt.title('age distribution')
plt.xlabel('age')
plt.ylabel('class')
plt.show()
```

```python
sns.displot(df, x='age', color='red', kind='kde')
plt.title('age distribution')
plt.xlabel('age')
plt.show()
```

```python
sns.displot(df, x='fare', y='age', hue='sex' , kind='hist')
plt.title('fare distribution')
plt.xlabel('fare')
plt.ylabel('age')
plt.show()
```

```python
sns.catplot(df, x='age', y='class', color='red')
plt.title('age distribution')
plt.xlabel('age')
plt.ylabel('class')
plt.show()
```
